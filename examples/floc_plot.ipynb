# Floc Plot

This notebook plots all of the floc proxy values

#### Setup environment

```python
%matplotlib inline
import pycamhd.pycamhd as camhd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 200
```

#### Load data

```python
frame_timestamp = []
floc_proxy = []
with open('d2_floc.csv', 'r') as csvfile:
    for line in csvfile:
        floc_proxy.append(int(line.split(',')[3].strip()))
        frame_timestamp.append(float(line.split(',')[2]))

with open('d3_floc.csv', 'r') as csvfile:
    for line in csvfile:
        floc_proxy.append(int(line.split(',')[3].strip()))
        frame_timestamp.append(float(line.split(',')[2]))
        
with open('d4_floc.csv', 'r') as csvfile:
    for line in csvfile:
        floc_proxy.append(int(line.split(',')[3].strip()))
        frame_timestamp.append(float(line.split(',')[2]))
```

```python
frame_timestamp[0]
```

#### Plot a two-dimensional multivariate histogram of the results

```python
plt.rc('font', size=11)
fig, ax = plt.subplots();
fig.set_size_inches(16, 6);
fig.frameon = False
hb1 = ax.hexbin(frame_timestamp, floc_proxy, vmin=0, vmax=2, bins='log', linewidths=0.25,
  gridsize=(450,4000), mincnt=3, cmap=plt.cm.BuPu)
fig.colorbar(hb1)
ax.set_ylim([0, 7000])
ax.set_xlim([frame_timestamp[0],frame_timestamp[-1]])
ax.yaxis.grid(True)
ax.xaxis.grid(True)
months = dates.MonthLocator()  # every month
monthsFmt = dates.DateFormatter('%b')
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(monthsFmt)
plt.ylabel('Floc Proxy Value');

# print
#fig_dpi = 300
#fig.savefig('floc_hexbin.png', bbox_inches='tight', transparent=True,
#  pad_inches=0, orientation='portrait', format='png', dpi=fig_dpi);
```
