# PyCamHD

This page shows some examples of how to use PyCamHD on the CamHDHub.

#### Setup your environment

```python
%matplotlib inline
import pycamhd as camhd
import numpy as np
import matplotlib.pyplot as plt
from numpngw import write_png
plt.rc('figure', figsize=(12, 12))
```

#### Load a frame using pycamhd

```python
filename = '/data/CAMHDA301-20160506T000000Z.mov'
frame_number = 7200
frame = camhd.get_frame(filename, frame_number, 'gray16le')
```

#### Convert the frame to an ndarray

```python
dt = np.dtype('uint16').newbyteorder('<')
framenp = np.frombuffer(frame.planes[0], dt).reshape(frame.height, frame.width)
```

#### Show the image

```python
imgplot = plt.imshow(framenp.squeeze(), cmap="gray")
```

#### Write the ndarray to a PNG
The write_png package seems to be prety smart about writing out PNG files based on the datatype of the ndarray.

```python
png_file = "/tmp/frame_%s.png" % frame_number # writes into /tmp
write_png(png_file, framenp)
```
