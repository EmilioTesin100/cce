# PyCamHD

This page shows some examples of how to use PyCamHD on the CamHDHub.

#### Setup your environment

```python
%matplotlib inline
import pycamhd as camhd
import numpy as np
import matplotlib.pyplot as plt
from numpngw import write_png
plt.rc('figure', figsize=(12, 12))
```

#### Load a frame using pycamhd

Returns an np.ndarray. Supports the following pixel formats: 'rgb24', 'bgr24', 'rgb48le', 'rgb48be', 'bgr48le', 'bgr48be', 'gray', 'gray16le', and 'gray16be'.

```python
filename = '/data/CAMHDA301-20160506T000000Z.mov'
frame_number = 7200
frame = camhd.get_frame(filename, frame_number, 'rgb24')
```

#### Show the image

```python
imgplot = plt.imshow(frame)
```

#### Write the ndarray to a PNG
The write_png package seems to be prety smart about writing out PNG files based on the datatype and shape of the ndarray.

```python
png_file = "/tmp/frame_%s.png" % frame_number # writes into /tmp
write_png(png_file, frame)
```
